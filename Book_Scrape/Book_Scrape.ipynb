```python
import requests
from bs4 import BeautifulSoup
import pandas as pd


response = requests.get('https://books.toscrape.com/catalogue/category/books/childrens_11/index.html')
response
```

```python
soup = BeautifulSoup(response.content, 'lxml')
soup
```

```python
soup.find('h3')
```

```python
names = soup.find_all('h3')
names
```

```python
names=[name.a.get('title') for name in names]
```

```python
prices =soup.find_all('p',{'class':'price_color'})
prices=[price.text for price in prices]
```

```python
links = soup.find_all('h3')
links = ['https://books.toscrape.com/catalogue/'+link.a['href'].replace('../','') for link in links]

links
```

```python
responses = [requests.get(f'{description}') for description in links]
soups = [BeautifulSoup(response.content, 'lxml') for response in responses ]
soups
```

```python
descriptions = [soup.select_one('#product_description + p').text.strip() if soup.select_one('#product_description + p') else "No description found" for soup in soups]

descriptions
```

```python
import pandas as pd
dic_book ={
    'Book name':names,
    'price' : prices,
    'links' :links,
    ' Descriptions':descriptions,
}
df = pd.DataFrame(dic_book  )
df
```

```python
df.to_csv('book_links.csv', index=False)
```
